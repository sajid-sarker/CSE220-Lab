Hashing & Hashtables

```python
class Student:
    def __init__(self, student_id, name, grade, next=None):
        self.student_id = student_id
        self.name = name
        self.grade = grade
        self.next = next

class StudentGradesDatabase:
    def __init__(self, size):
        self.pages = [None] * size

    # do by yourself
    def find_pageNumber(self, key):
      s = str(key)
      c = 0
      for i in range(len(s)):
        c += int(s[i])
      return c % len(self.pages)

    def add_student(self, student_id, name, grade):
        if self.find_student(student_id) == 'Found':
            print(student_id, 'Already added')
            print('===============================')
            return

        key = self.find_pageNumber(student_id)
        node = Student(student_id, name, grade)
        if self.pages[key] is None:
            self.pages[key] = node
        else:
            node.next = self.pages[key]
            self.pages[key] = node

    # do by yourself
    def get_student_grades(self, student_id):
        k = self.find_pageNumber(student_id)
        if self.find_student(student_id) == 'Found':
          t = self.pages[k]
          while t != None:
            if t.student_id == student_id:
              return t.grade
            t = t.next
        else:
          return None

    def find_student(self, id):
        key = self.find_pageNumber(id)
        temp = self.pages[key]
        while temp is not None:
            if temp.student_id == id:
                return 'Found'
            temp = temp.next
        return 'Not found'

    # do by yourself
    def printAll(self):
      for i in range(len(self.pages)):
        print("Page No:", i+1)
        print('=========')
        t = self.pages[i]
        while t!= None:
          print('Student ID:', t.student_id)
          print('Name:', t.name)
          print('Grade:', t.grade)
          t = t.next
        print('=========')
```

```python
db = StudentGradesDatabase(size = 5)  # Choose an appropriate size for the hashtable

db.add_student(101, "Barry", 3.70)
db.add_student(102, "Wells", 4.00)
grade = db.get_student_grades("103")
print("Student Found and grade = :", grade) if grade is not None else print("Student with given id not found")
# output: Student with given id not found


db.add_student(103, "Thawne", 3.99)
db.add_student(104, "Zolomon", 3.50)
db.add_student(105, "Grodd", 3.75)
db.add_student(106, "Cisco", 4.00)
db.add_student(107, "Caitlin", 3.98)
db.add_student(100, "Cicada", 3.40)

db.printAll()
'''
output:
Page No: 1
=========
Student ID: 104
Name: Zolomon
Grade: 3.5
=========
Page No: 2
=========
Student ID: 100
Name: Cicada
Grade: 3.4
Student ID: 105
Name: Grodd
Grade: 3.75
=========
Page No: 3
=========
Student ID: 106
Name: Cisco
Grade: 4.0
Student ID: 101
Name: Barry
Grade: 3.7
=========
Page No: 4
=========
Student ID: 107
Name: Caitlin
Grade: 3.98
Student ID: 102
Name: Wells
Grade: 4.0
=========
Page No: 5
=========
Student ID: 103
Name: Thawne
Grade: 3.99
=========
'''
grade = db.get_student_grades(103)
print("Student Found and grade = ", grade) if grade is not None else print("Student with given id not found")
# output: Student Found and grade =  3.99
```
